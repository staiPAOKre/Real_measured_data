```python
import matplotlib.pyplot as plt

# Define a function to create histograms for each column
def create_hist(df, x_title, y_title):
    num_cols = df.shape[1]
    fig, axs = plt.subplots(1, num_cols, figsize=(12, 4))
    fig.subplots_adjust(hspace=0.5, wspace=0.5)
    for i in range(num_cols):
        if i == 0:
            axs[i].set_title('X_Axis')
            axs[i].grid(True)
        elif i == 1:
            axs[i].set_title('Y_Axis')
            axs[i].grid(True)
        elif i == 2:
            axs[i].set_title('Z_Axis')
            axs[i].grid(True)
        else:
            axs[i].set_title('RPM')
        axs[i].hist(df.iloc[:, i])
        axs[i].set_xlabel(x_title)
        axs[i].set_ylabel(y_title)
    axs[i].grid(True)
    plt.tight_layout()
    plt.show()

# Call the function for each dataframe
create_hist(norm_values, 'm/s^2', 'Count')
create_hist(norm_values1, 'm/s^2', 'Count')
create_hist(norm_values2, 'm/s^2', 'Count')
create_hist(norm_values3, 'm/s^2', 'Count')
```

```python
import numpy as np

def shuffle_and_split_data(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    data_without_rpm = data.iloc[:, :-1]  # drop the last column (RPM)
    train_data = data_without_rpm.iloc[train_indices]
    test_data = data_without_rpm.iloc[test_indices]
    return train_data, test_data
```

```python
# Shuffle and split each dataframe into training and testing sets
data_train, data_test = shuffle_and_split_data(norm_values, test_ratio=0.2)
data1_train, data1_test = shuffle_and_split_data(norm_values1, test_ratio=0.2)
data2_train, data2_test = shuffle_and_split_data(norm_values2, test_ratio=0.2)
data3_train, data3_test = shuffle_and_split_data(norm_values3, test_ratio=0.2)
```

```python
from zlib import crc32

def is_id_in_test_set(identifier, test_ratio):
    return crc32(np.int64(identifier)) < test_ratio * 2**32

def split_data_with_id_hash(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: is_id_in_test_set(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam


train_data = np.random.standard_normal((1000,4))
train_labels = np.random.standard_normal((1000,1))

model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(4,)))
model.add(Dense(64, activation='relu'))
#model.add(Dense(32, activation='relu'))
model.add(Dense(2, activation='linear'))

optimizer = Adam(learning_rate=0.0001)

model.compile(loss='huber_loss', optimizer=optimizer, metrics=['mse', 'mae', 'accuracy'])
history = model.fit(train_data, train_labels, epochs=100, batch_size=32, validation_split=0.1)
```
